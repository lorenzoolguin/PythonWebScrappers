```python
#payload = {"statisticCategory":statisticCategory.upper(),"seasonType":"REG","d-447263-p":str(p),"season":season}
url = 'https://www.nfl.com/stats/player-stats/category/passing/2020/POST/all/passingyards/desc'
response = requests.get(url)
```

```python
print("Response:", response.status_code,response.url) #200 means it went through
```

```python
soup = BeautifulSoup(response.text,'html.parser') #html.parser or lxml, lxml can handle broken html code. (thats what some guy said)
headers_list = []
headers = soup.find_all('th')
for col_header in headers:
    headers_list.append(col_header.text)
    #print(col_header.text)
headers_list = [col_header.strip('\n') for col_header in headers_list]
headers_list
```

```python

data_list = []
data = soup.find_all('td')
for player_data in data:
    data_list.append(player_data.text)
    #print(player_data.text)
data_list = [player_data.strip('\n') and player_data.strip() for player_data in data_list]
data_list

#list of lists using by a step value here denoted via K.
k = len(headers_list)
data_final = [] 
for i in range(0, len(data_list), k):  
    data_final.append(data_list[i:k])
    k += len(headers_list)
data_final
```
